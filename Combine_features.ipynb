```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sample_order_prior_user_product = pd.read_csv("instacart/sample_order_prior_user_product.csv")
sample_order_prior_user_product
```

```python
sample_order_prior_user_product['times_reordered'] = sample_order_prior_user_product['num_prod_bought']-1
```

```python
sample_order_prior_user_department = pd.read_csv("instacart/sample_order_prior_user_department.csv")
sample_order_prior_user_aisle = pd.read_csv("instacart/sample_order_prior_user_aisle.csv")
```

```python
sample_order_prior_user_department
```

```python
sample_order_prior_user_aisle
```

# Ratio of product/order bought correspond to department/aisle

## Merge product, department, aisle

```python
sample_order_prior_user_product = sample_order_prior_user_product.merge(sample_order_prior_user_department, how='left', on=['user_id', 'department_id'])
sample_order_prior_user_product = sample_order_prior_user_product.merge(sample_order_prior_user_aisle, how='left', on=['user_id', 'aisle_id'])
sample_order_prior_user_product
```

## Calculate ratio

```python
sample_order_prior_user_product['ratio_prod_bought_dep'] = sample_order_prior_user_product['num_prod_bought']/sample_order_prior_user_product['num_prod_bought_dep']
sample_order_prior_user_product['ratio_prod_bought_ais'] = sample_order_prior_user_product['num_prod_bought']/sample_order_prior_user_product['num_prod_bought_ais']
sample_order_prior_user_product['ratio_order_bought_dep'] = sample_order_prior_user_product['num_prod_bought']/sample_order_prior_user_product['num_order_bought_dep']
sample_order_prior_user_product['ratio_order_bought_ais'] = sample_order_prior_user_product['num_prod_bought']/sample_order_prior_user_product['num_order_bought_ais']
```

```python
sample_order_prior_user_product
```

# Number of product bought

## Total reordered products

```python
reordered_user_product = sample_order_prior_user_product[sample_order_prior_user_product['num_prod_bought']>1]
reordered_user_product = reordered_user_product.groupby('user_id').agg(
    num_dist_prod_reordered=('product_id', 'count'),
    num_total_prod_reordered=('times_reordered', 'sum')
)
reordered_user_product.reset_index(inplace=True)
reordered_user_product
```

## Total products bought

```python
num_prod_bought = sample_order_prior_user_product.groupby('user_id').agg(
    num_distinct_prod_user_total=('product_id', 'count'),
    num_prod_user_total=('num_prod_bought', sum)
).reset_index()
num_prod_bought
```

## Merge

```python
sample_order_prior_user_product = sample_order_prior_user_product.merge(reordered_user_product, how='left', on=['user_id'])
sample_order_prior_user_product = sample_order_prior_user_product.merge(num_prod_bought, how='left', on=['user_id'])
sample_order_prior_user_product
```

```python
sample_order_prior_user_product['ratio_dist_prod_reordered'] = sample_order_prior_user_product['num_dist_prod_reordered']/sample_order_prior_user_product['num_distinct_prod_user_total']
sample_order_prior_user_product
```

# Combine with user features

```python
sample_order_prior_user = pd.read_csv("instacart/sample_order_prior_user.csv")
sample_order_prior_user.rename(columns={'days_cumsum_last':'days_cumsum_user'}, inplace=True)
sample_order_prior_user
```

```python
final_df = sample_order_prior_user_product.merge(sample_order_prior_user, how='left', on='user_id')
final_df
```

```python
final_df.columns
```

## Ratio of order bought

```python
final_df['ratio_order_bought'] = final_df['num_prod_bought']/final_df['number_user_order']
final_df
```

## Number of products per order

```python
final_df['num_user_prod_per_order'] = final_df['num_prod_user_total']/final_df['number_user_order']
final_df
```

## Ratio of reordered products/basket

```python
final_df['reordered_prod_per_order'] = final_df['num_total_prod_reordered']/final_df['num_prod_user_total']
final_df
```

# Summarize

```python
final_df.info()
```

## column name meanings
- times_reordered = number of time this product has been reordered
- num_total_prod_reordered = sum of all products reordered (not distinct)
- ratio_dist_prod_reordered = number of distinct reordered products / number of total distinct products
- ratio_order_bought = number of time this product bought / number of total order
- num_user_prod_per_order = number of total products bought / number of total order = average number of products per order
- reordered_prod_per_order = number of total reordered products bought / number of total products bought = average proportion of reordered items to all items in an order
