## Find time different

```python
train_data['order_dow'].unique()
```

```python
def diff_dow(a,b):
    if(abs(a-b)<=3):
        return abs(a-b)
    return 7-abs(a-b)
```

```python
diff_dow(1,6)
```

```python
train_data['dow_prod_usually_diff'] = train_data.apply(lambda x: diff_dow(x.dow_usually_bought, x.order_dow), axis=1)
train_data['dow_user_usually_diff'] = train_data.apply(lambda x: diff_dow(x.dow_user_usually_bought, x.order_dow), axis=1)
train_data
```

```python
train_data['hod_prod_usually_diff'] = (train_data['hod_usually_bought'] - train_data['order_hour_of_day']).abs()
train_data['hod_user_usually_diff'] = (train_data['hod_user_usually_bought'] - train_data['order_hour_of_day']).abs()
train_data
```

```python
train_data['days_since_prod_bought'] = train_data['days_cumsum_order'] - train_data['days_cumsum']
train_data
```

```python
train_data['prod_bought_period_diff_mean'] = (train_data['days_since_prod_bought'] - train_data['mean_days_prod_bought_period']).abs()
train_data['prod_bought_period_diff_user'] = (train_data['days_since_prod_bought'] - train_data['average_days_period']).abs()
train_data['dow_prod_usually_diff_mean'] = (train_data['order_dow'] - train_data['mean_dow_prod_usually_bought']).abs()
train_data['hod_prod_usually_diff_mean'] = (train_data['order_hour_of_day'] - train_data['mean_hod_prod_usually_bought']).abs()
```

```python
train_data['average_day_period_diff'] = (train_data['average_daydiff_user'] - train_data['days_since_prior_order']).abs()
```

# Summarize

This dataset is trying to capture the behaviour of each pair (user, product_id).<br>
It also consider other people's behaviour on the same product.<br>
When we use on test set, we will predict on each (user, product_id) of all the products the user bought before.

- 'user_id', 
- 'product_id', 
- 'times_reordered' = how many time user reorder this product
- 'ratio_prod_bought_dep' = ratio of this product bought / total product in the same department
- 'ratio_prod_bought_ais',
- 'ratio_order_bought_dep' = ratio of orders contain this product / total orders contain products in the same department 
- 'ratio_order_bought_ais', 
- 'ratio_dist_prod_reordered' = ratio of distinct products that have been reordered
- 'ratio_order_bought' = number of orders contain this item / total orders 
- 'num_user_prod_per_order' = average number of products in an order of this user
- 'reordered_prod_per_order' = proportion of reordered product in an order of this user 
- 'dow_prod_usually_diff' = different in dow that this user usually buy this product 
- 'dow_user_usually_diff' = different in dow that this user usually make an order
- 'hod_prod_usually_diff' = different in hod that this user usually by this product 
- 'hod_user_usually_diff' = different in hod that this user usually make an order
- 'prod_bought_period_diff_mean' = different in period that this user usually buy this product compared to other people
- 'prod_bought_period_diff_user' = different in period that this user usually buy this product compared to himself
- 'dow_prod_usually_diff_mean' = different in dow that this user buy this product compared to other people
- 'hod_prod_usually_diff_mean' = different in hod that this user buy this product compared to other people
- 'average_day_period_diff' = different in period that this user usually make an order
- 'reordered' = label
